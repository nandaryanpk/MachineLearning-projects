```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# soal nomor 1
curve = []
for x in range(0,10):
    y = np.log(x+5)
    curve.append(y)
plt.xlabel("x:0-10")
plt.plot(curve);
```

```python
# soal nomor 4
curve2 = []
for x in range(0,10):
    y = x**2 + 4*x +3
    curve2.append(y)
plt.xlabel("x:0-10")
plt.plot(curve2);
```
